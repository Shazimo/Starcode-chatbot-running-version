## Variablen

Dazu speichern wir ihn in eine Variable.
Eine Variable ist wie eine beschriftete Kiste.
Sie hat eine Beschriftung bzw. einen Namen.
Und sie hat einen Inhalt: ein Wort, einen Text oder eine Zahl  
Beispiel:  

zahl1 = 3  
zahl2 = 7.5  
wort1 = "Blume"  

````python
zahl1 = 3 
````

````python
print(zahl1)
````
